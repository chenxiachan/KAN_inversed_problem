```python
from graphviz import Digraph
from IPython.display import display

# 创建图
dot = Digraph(comment='KAN vs ANN Decision Tree')
dot.attr(rankdir='TB', size='28')

# 定义节点样式
dot.attr('node', shape='diamond')
dot.attr('edge', color='blue')

# 创建节点
dot.node('A', 'Data Availability')
dot.node('B', 'Problem Complexity')
dot.node('C', 'Interpretability\nRequirements')
dot.node('D', 'Dynamic Nature\nof Problem')
dot.node('E', 'Extreme Value\nPrediction')
dot.node('F', 'Transfer Learning /\nGeneralization Needs')
dot.node('G', 'Computational\nResources')
dot.node('H', 'Need for\nContinuous Learning')
dot.node('I', 'Discovering Underlying\nRelationships')
dot.node('J', 'Real-time\nPerformance')

# 创建叶节点（矩形）
dot.attr('node', shape='rectangle', style='filled')
dot.node('KAN', 'KAN', color='lightgreen')
dot.node('ANN', 'ANN', color='lightblue')
dot.node('HYBRID', 'Hybrid Approach', color='lightyellow')
dot.node('REASSESS', 'Reassess with\nData Augmentation', color='lightgrey')

# 创建边
dot.edge('A', 'KAN', 'Limited\nNeed Interpretability')
dot.edge('A', 'REASSESS', 'Limited\nNo Interpretability Need')
dot.edge('A', 'B', 'Abundant')
dot.edge('B', 'KAN', 'Simple')
dot.edge('B', 'C', 'Complex')
dot.edge('C', 'KAN', 'High')
dot.edge('C', 'D', 'Low to Medium')
dot.edge('D', 'KAN', 'Highly Dynamic')
dot.edge('D', 'E', 'Stable')
dot.edge('E', 'KAN', 'Critical')
dot.edge('E', 'F', 'Not Critical')
dot.edge('F', 'ANN', 'High')
dot.edge('F', 'G', 'Low')
dot.edge('G', 'ANN', 'Limited')
dot.edge('G', 'H', 'Abundant')
dot.edge('H', 'HYBRID', 'High')
dot.edge('H', 'I', 'Low')
dot.edge('I', 'KAN', 'High')
dot.edge('I', 'J', 'Low')
dot.edge('J', 'ANN', 'Critical')
dot.edge('J', 'KAN', 'Not Critical')

# 在 Jupyter 中显示图
display(dot)

# 保存图为 PNG 文件
dot.render('kan_ann_decision_tree', format='png', cleanup=True)
print("Decision tree image has been generated as 'kan_ann_decision_tree.png'")
```
